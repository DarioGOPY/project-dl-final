# Proyecto Final - Aprendizaje Profundo Avanzado

## Guía de proyecto ML para portfolio — CNN, GANs, RAGs, LSTM, Transformers 

Esta guía paso a paso está dada para la construcción de un proyecto de portfolio que incluya CNN (clasificación de imágenes), GANs (generación de imágenes), RAGs (Retrieval-Augmented Generation para tareas de texto conversacional o QA), LSTM (modelado de secuencias) y Transformers (tareas de NLP: clasificación / fine-tuning).  

El entregable final será un repositorio de GitHub con notebooks en Jupyter, código reproducible, documentación (README + reporte técnico).  

Para esto hemos creado el repositorio publico: https://github.com/DarioGOPY/project-dl-final#

## Objetivos del proyecto 

Implementar y comparar modelos concretos de cada familia (CNN, GAN, LSTM, Transformer) en tareas representativas. 
Preparar datos, entrenar, evaluar y explicar resultados. 
Documentar el proceso en notebooks y un reporte profesional que sirva para entrevistas. 
Practicar buenas prácticas: control de versiones, experiment tracking, reproducibilidad y visualización. 

## Estrategia propuesta de desarrollo V1:
+ Como la mayoria de los cuadernos requieren GPU y en algunos casos como GANs es imprecindible vamos a necesitar usar google colab con la GPU T4.
+ Para agotar los creditos la estrategia es preparar los cuadernos en forma local hasta validar que se ejecutan sin errores al menos hasta la primer epoca, una vez hecho esto ejecutamos el cuaderno completo en la GPU T4 usando Google Colab.
